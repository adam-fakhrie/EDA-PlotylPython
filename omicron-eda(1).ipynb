# **Memuat Data**

**Merubah Tipe Data pada 'date' menjadi DateTime**

```python
df_olah['date'] = pd.to_datetime(df_olah['date'])
df_olah.info()
```

# **Visualisasi**

**Memuat Library**

```python
import plotly.express as px
from plotly.offline import plot, iplot, init_notebook_mode #agar tidak terputus dengan library ketika shell di proses
init_notebook_mode(connected=True)

#import matplotlib.pyplot as plt
#%matplotlib inline
#import seaborn as sns

print("Setup Complete") #memastikan agar proses sudah benar berjalan
```

```python
fig = px.histogram(df_olah, x="date", y="num_sequences", color="variant", title="Jumlah Kasus dalam Data per Tahun dengan Histogram")
fig.show()
```

```python
jmlh = df_olah.groupby(['variant', 'location'])['num_sequences'].sum().reset_index()

fig = px.pie(jmlh, values='num_sequences', names='variant', 
            title="Presentase Jumlah Kasus dalam Data dengan Pie Chart")
fig.show()
```

```python
olah_shw = px.bar(jmlh, x = 'variant', y = 'num_sequences', color='location', title="Varian Kasus Terbanyak per Negara dengan Bar Chart")
olah_shw.update_layout(    
    xaxis_title="Variant",
    yaxis_title="Total Sequences"
)
olah_shw.show()
```

```python
df = df_olah[df_data["variant"] == "Omicron"]

var2 = df.groupby(['variant', 'location'])['num_sequences'].sum().reset_index()

fig = px.scatter(x=var2["num_sequences"], y=var2["num_sequences"],
                 size=var2["num_sequences"], color=var2["location"],
                 hover_name=var2["location"], log_x=True, size_max=60, title="Visual Penyebaran Virus Omicron dengan Bubble Chart")
fig.show()
```

```python
fig = px.scatter(df_olah, x="date", y="num_sequences", color="location", facet_col="variant", 
                 #size="num_sequences", 
                 title="Visual Data Keseluruhan dengan Scatter Plots")
fig.show()
```
